````python
fig, axes = plt.subplots(4, 4)
for x, y, ax in zip(digits.data, digits.target, axes.ravel()):
    ax.set_title(y)
    ax.imshow(x.reshape(8, 8), cmap="gray_r")
    ax.set_xticks(())
    ax.set_yticks(())
plt.tight_layout()
````

````python
import sklearn.datasets
import os
iris_path = os.path.join(sklearn.datasets.__path__[0], 'data', 'iris.csv')

import pandas as pd
iris = pd.read_csv(iris_path, skiprows=1, header=None, names=['sepal_length','sepal_width','petal_length','petal_width','class'])
data = iris.drop('class', axis=1) #.as_matrix()  # This returns a numpy array
target = iris['class']
print(type(data))
print(type(target))
````

````python
plt.figure()
#plt.scatter(data[:,0], data[:,1], c=target) # To use if a numpy array is used
plt.scatter(data['sepal_length'], data['sepal_width'], c=target);
plt.figure()
#plt.scatter(data[:,2], data[:,3], c=target) # To use if a numpy array is used
plt.scatter(data['petal_length'], data['petal_width'], c=target);
````
